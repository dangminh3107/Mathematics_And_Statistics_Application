#### Bài 1:  Romeo và Juliet hẹn gặp nhau từ 8h đến 9h. Giả sử thời điểm đến của mỗi người là ngẫu nhiên trong khoảng thời gian đó, hơn nữa, người tới sớm sẽ chờ người tới sau không quá 20 phút. Không gặp thì chia tay. Tính xác suất chia tay? Khảo sát xác suất chia tay theo thời gian chờ $t$ phút $(0 \le t \le 60).$

### Giải:
Gọi A là biến có Romeo và Juliet không chia tay
* Gọi $x$ là thời điểm Romeo đến ($0 \le x \le 60$)
* Gọi $y$ là thời điểm Romeo đến ($0 \le y \le 60$)
Ta có không gian mẫu : $ \Omega = \{ (x,y) : 0 \le x \le 60  , \le y \le 60\} , |\Omega| = S_{\text{vuông}} = 60^2 $
* Nếu biểu diễn $x, y$ lần lượt Ox và Oy, thời điểm đến của 2 người được biểu diễn thành cặp điểm $(x,y)$ nằm trong hình vuông có cạnh là $60$


* Để 2 người gặp nhau thì thời gian đến $x$ và $y$ phải thoã điều kiện :\
$$ | x - y | \le 20 $$
* Các cặp điểm (x,y) nằm trong đa giác $OABCDE$ là các điểm thời gian mà Julio và Romelet đến và gặp nhau
* Theo công thức xác suất hình học ta có : 

$$ P(A) = \frac{S_{OABCDE}}{S_{vuông}} = \frac{60^2 - 40^2}{60^2} = \frac{5}{9} = 0.(5) = 0.555555555555 $$

Vậy xác suất Romelet và Julio chia tay là : $$ P(\overline{A}) = 1 - P(C) = 1 - \frac{5}{9} = 0.44444444 $$

#### Bài 2: Bộ bài tiến lên 52 lá, xào “đều”, chia cho 4 người chơi. Khả năng có người được tứ quí heo là hiếm cỡ nào?

### Giải:
* Gọi $B$ là biến cố nhận được Tứ quý heo khi đánh bài tiến lên
* Ta có không gian mẫu là số cách người chơi nhận được 13 lá bài ngẫu nhiên theo lượt chia \
$$ | \Omega | = \binom{52}{13} $$
Vì có duy nhất một trường hợp tứ quý heo, nên số cách nhận các lá bài còn lại là $ \binom{49}{9} $
\
$$P(B) = \frac{1*\binom{49}{9}}{\binom{52}{13}} = \frac{11}{3400} = 0.00323523 $$
