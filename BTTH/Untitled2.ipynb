### 2. Quản lý Package
   - Liệt kê package : pip list \
   - Tìm kiếm package: pip3 search < module_name > \
   - Cài đặt package: pip3 install < module_name > \
   - Cập nhật: pip3 install --upgrade < module_name >==< module_version > \
   - Gỡ bỏ package: pip3 uninstall < module_name >

```python
try:
    fin = open('name.txt','r')
    s = fin.read()
    fout = open('output.txt','w')
    fout.write(s)
except IOError:
    print("File not found")
finally:
    fin.close()
    fout.close()
```
