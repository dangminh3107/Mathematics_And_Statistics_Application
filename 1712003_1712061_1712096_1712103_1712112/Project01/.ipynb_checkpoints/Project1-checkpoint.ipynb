```python
# Import thư viện
import csv
import numpy as np
import math
import pickle
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
# convert data to number

# Trả về dictionary ứng với mõi filename (file.csv), dictionary có giá trị {"9 - 3":1, "9 - 5":2}
# trong đó: - "3 - 1" là giá trị lấy từ file csv
#           - 1 là thứ tự xuất hiện của "3 - 1" trong file csv ứng với giá trung bình của "3 - 1" là thấp nhất
def convertSortedDataToDictionary(filename):
    filecsv = filename
    dictionary = {}
    with open(filecsv) as file:
        r = csv.reader(file)
        i = 1
        for line in r:
            dictionary[str(line[0])] = i
            i = i + 1
    return dictionary

model_a_z = convertSortedDataToDictionary('predata\\model_a_z.csv')
brand_a_z = convertSortedDataToDictionary('predata\\brand_a_z.csv')
color_a_z = convertSortedDataToDictionary('predata\\color_a_z.csv')
bodytype_a_z = convertSortedDataToDictionary('predata\\bodytype_a_z.csv')
# xxx_a_z.csv là các file sắp xếp các phần tử xxx theo thứ tự tăng dần giá trung bình của từng xxx
# Ví dụ như model_a_z.csv có 3 phần tử đầu là a, b, c tức a có trung bình giá của các xe nhỏ nhất

# hàm sinh index của phần tử phân biệt trong mãng
# trả về dictionary thể hiện thứ tự phần tử phân biệt s trong model_dic
def dic(model_dic, s, i):
    if (s in model_dic[i]):
        index = model_dic[i].index(s)
        return index
    model_dic[i].append(s)
    return (len(model_dic[i]))
```

```python
w, h = 8, 30000;
model_dic = [[0 for x in range(w)] for y in range(h)]

# đọc và convert data từ file X_Train.csv, Y_Train.csv sang X, Y
def readData(x_file, y_file, isTrain):
    X = []
    y = []
    a = []
    index = []
    
    with open(x_file) as f:
        r = csv.reader(f)
        # Skip header
        next(r)
        # Đọc X, y
        for line in r:
            xline = [1.0]
            i = 0
            # Đọc từ 2 trở về sau bỏ cột đầu là thứ tự
            # Các cột từ 1 - 12 thể hiện các thông tin về xe
            for s in line[1:13]:
                if (i!=-11):
                    if (s =="" or s == "False"):
                        if(s == ""): xline.append(4.0)
                        if(s == "False"): xline.append(float(0))
                    else:
                        if (s =="True"):
                            xline.append(float(1))
                        else: 
                            if (s.isdigit()):
                                # Years
                                if (i==5):
                                    a = float(s)
                                    a=(a - 1990)**2
                                    xline.append(float(a))
                                else:
                                    # odo
                                    if(i == 4):
                                        a = float(s)
                                        xline.append(a)    
                                    else:
                                        if(i != 11):
                                            xline.append(float(s))
                                        else:
                                            a = float(s)
                                            #if(a > 40):
                                            #    a = 40
                                            a = a**2
                                            xline.append(a)
                                        
                            else:
                                # Xử lí các trường hợp gắn nhãn category
                                isCheck = [0, 1, 3, 9, 10, 6, 7]
                                #print(s)
                                if(i not in isCheck): 
                                    xline.append(float(dic(model_dic, s, i)))
                                else:
                                    if(i == 0): #manufacturer
                                        if(s in brand_a_z):
                                            xline.append(float(brand_a_z[s]**3))
                                        else:
                                            xline.append(float(1))
                                    if(i == 1): #model
                                        if(s in model_a_z):
                                            xline.append(float(model_a_z[s]**5))
                                        else:
                                            xline.append(float(1))
                                    if(i == 9): #bodyType
                                        if(s in bodytype_a_z):
                                            xline.append(float(bodytype_a_z[s]**10))
                                        else:
                                            xline.append(float(1))
                                    if(i == 3): #color
                                        #continue
                                        if(s in color_a_z):
                                            xline.append(float(color_a_z[s]))
                                        else:
                                            xline.append(float(1))
                                    if(i == 10): #drivetrain
                                        listKey = ['all', 'rear', 'front']
                                        if(s in  listKey):
                                            if(s == 'all'):
                                                xline.append(float(25))
                                            if(s == 'rear'):
                                                    xline.append(float(15))
                                            if(s == 'front'):
                                                xline.append(float(1))
                                        else: 
                                            xline.append(float(1))
                                    if(i==7): #engineType
                                        listKey = ['diesel', 'electric', 'gasoline']
                                        if(s in listKey):
                                            if(s == 'diesel'):
                                                xline.append(float(7))
                                            if(s == 'electric'):
                                                xline.append(float(25))
                                            if(s == 'gasoline'):
                                                xline.append(float(5))
                                        else:
                                            xline.append(float(5))
                                    if(i==6): #engineFuel
                                        listKey = ['diesel', 'electric', 'gasoline', 'hybrid-diesel', 'hybrid-petrol']
                                        if(s in listKey):
                                            if(s == 'diesel'):
                                                xline.append(float(7))
                                            if(s == 'electric'):
                                                xline.append(float(15))
                                            if(s == 'gasoline'):
                                                xline.append(float(5))
                                            if(s == 'hybrid-diesel'):
                                                xline.append(float(5))    
                                            if(s == 'hybrid-petrol'):
                                                xline.append(float(10))
                                        else:
                                            xline.append(float(5))

                                        
                i+=1
                
            # chú ý, feature bất đầu từ 13
            # Các cột bắt đầu từ 13 là thể hiện xe có các tính năng nào
            # Khi train thực hiện bỏ cột 13, cột 18 (i != 17)
            for s in line[14:]:
                if(i != 17):
                    if ((s =="" or s == "False")):
                        xline.append(float(0))
                    else:
                        if (s =="True"):
                            xline.append(float(1))
                        else: 
                            if (s.isdigit()):
                                xline.append(float(s))
                            else:
                                # Xử lí các trường hợp ngoại lai
                                xline.append(float(dic(model_dic, s, i)))
                i+=1
            X.append(xline)
            
        # Đọc y vào là cột
        with open(y_file) as fy:
            r = csv.reader(fy)
            # Skip header
            next(r)
            # Đọc X, y
            for line in r:
                xline = [1.0]
                # Đọc từ 2 trở về sau bỏ cột đầu là thứ tự và cột tiếp theo là y (GPA)
                for s in line[1:]:
                    if (s ==""):
                        y.append(float(0))
                    else:
                        a = float(s)
                        if(isTrain): 
                            if(a < 200): a = 200
                        #Thực hiện loại bỏ các phần tử "giá (price)" vô lý.
                        #VD Xe có giá trị < 200$
                        #Do việc xử lí này ảnh hưởng đến giá trị dự đoán nên
                        y.append(float(a))
    return (X,y)
```

```python
# Nhập tên file thích hợp

# File X train
X_train = "data\\X_train.csv"

# File y train
Y_train = "data\\Y_train.csv"

# File x test
X_test = "data\\X_train.csv"

# File y test
Y_test = "data\\Y_train.csv"

# Các giá trị được sử dụng
# Tập x train
X_features_train = []

# Tập y train
Y_featues_train = []

# Tập x test
X_features_test = []

# Tập y test
Y_featues_test = []
```

### Lưu ý, đoạn code dưới đây bổ trợ cho quá trình test với tập train, không phải tập test thực
### Có 4 đoạn code phía sau thể hiển:
    - Chuẩn bị dữ liệu cho việc train và test bằng tập train
    - Train và test bằng thư viện Sklearn (file test được tách từ 10% file train)
    - Train và test bằng thủ công (file test được tách từ 10% file train)

### Khi muốn thực hiện với test thực, ta sửa lại các tham số:
-(1): x1, y1 = readData(X_train, Y_train, isTrain = False) - > x1, y1 = readData(X_test, Y_test, isTrain = False)

-(2): Xóa hết từ dòng: # Dòng này tách test input thành test input và train input

```python
# Đọc dữ liệu
x0, y0 = readData(X_train, Y_train, isTrain = True)
x1, y1 = readData(X_train, Y_train, isTrain = False)

# Convert thành định dạng phù hợp
X_features_train, Y_features_train = np.array(x0), np.array(y0) # train
X_features_test, Y_features_test = np.array(x1), np.array(y1)   # test

d = len(X_features_train) - 3000

# Dòng này tách test input thành test input và train input
X_features_train, Y_features_train = np.array(x0[:d]), np.array(y0[:d])
X_features_test, Y_features_test = np.array(x1[d:]), np.array(y1[d:])
```

### TRAIN và Test bằng Sklearn 

```python
# Thực hiện train và test bằng Sklearn# Train

# Train
reg = linear_model.LinearRegression()
reg.fit(X_features_train, Y_features_train)

# In kết quả dự đoán
i = 1
for data,actual in zip(X_features_test, Y_features_test):
    prediction = reg.predict([data])
    print('STT ' + str(i) + ': Prediction = '+str(prediction[0])+' -- Actual = '+str(actual))
    i+=1


# In RMSE
y_true = Y_features_test
y_pred = reg.predict(X_features_test)
rmse = mean_squared_error(y_true, y_pred)
print("RMSE:", math.sqrt(rmse))
# In giá trị R^2
print("R^2 :", reg.score(X_features_test, Y_features_test))
```

## Train và Test bằng mô hình

```python
X = X_features_train # Các vector dùng để train
y = np.transpose(Y_features_train) # Giá trị thực của xe trong lúc train

# Train bằng giải Beta
Xt = np.transpose(X)
XtX = np.dot(Xt,X) # X.T*X
Xty = np.dot(Xt,y) # X.T*y
beta = np.linalg.solve(XtX,Xty) ## Solve (X.T*X)Beta = X.T*y)
# convert sang mảng
z = []
for i in beta:
    z.append([i])
beta = np.array(z)

# In mô hình ra màn hình
print("B:")
print(beta)

# In kết quả dự đoán
i = 1
for data,actual in zip(X_features_test, Y_features_test):
    prediction = np.dot([data], beta)
    print('STT ' + str(i) + ': Prediction = '+str(prediction[0])+' -- Actual = '+str(actual))
    i+=1
    
# In giá trị RMSE và R-squares:
k = 21 # số cột của data
n = len(Y_features_test)# số hàng
mean = np.mean(Y_features_test) # tính mean
e = [] #rss sai số (chưa được giải thích bởi mô hình)
yy = [] # ess (dao động được giải thích bởi mô hình)
tss = []

# Tính các tổng các giao động
for data,actual in zip(X_features_test, Y_features_test):
    prediction = np.dot([data], beta)
    e.append((actual-(prediction[0][0]))**2)
    yy.append(((prediction[0][0]) - mean)**2)
    tss.append((actual - mean)**2)

ESS = np.sum(yy)
RSS = np.sum(e)
TSS = np.sum(tss)
R_Square = 1 - RSS/TSS
Adjusted_R_Square = 1 - (n-1)/(n-k)*(1 - R_Square)
print("RMSE: ", math.sqrt(RSS/n))
print("R_Square: ",R_Square)
print("Adjusted R_Square: ", Adjusted_R_Square)
```

```python
def train_manual(X_file, Y_file):
    #predata
    X, Y = readData(X_file, Y_file, isTrain = True)
    x0 = np.array(X)
    y0 = np.transpose(np.array(Y))
    
    # Train bằng giải Beta
    Xt = np.transpose(X)
    XtX = np.dot(Xt,x0) # X.T*X
    Xty = np.dot(Xt,y0) # X.T*y
    beta = np.linalg.solve(XtX,Xty) ## Solve (X.T*X)Beta = X.T*y)
    
    # convert sang mảng
    z = []
    for i in beta:
        z.append([i])
    beta = np.array(z)
    
    # Save model
    with open('beta.binary', 'wb') as beta_file:
        pickle.dump(beta, beta_file)
    return beta

def test_manual(X_file, Y_file):
    # Đọc beta
    beta = []
    with open('beta.binary', 'rb') as beta_file:
        beta = pickle.load(beta_file)
    
    # Đọc file test
    X_test_manual, Y_test_manual = readData(X_file, Y_file, isTrain = False)
    
    X_test_manual = np.array(X_test_manual)
    Y_test_manual = np.array(Y_test_manual)
    
    n = len(X_test_manual)
    # Dự đoán
    i = 1
    for data,actual in zip(X_test_manual, Y_test_manual):
        prediction = np.dot([data], beta)
        print('STT ' + str(i) + ': Prediction = '+str(prediction[0])+' -- Actual = '+str(actual))
        i+=1
    
    
    # Trả về kết quả RMSE
    # Tính các tổng các giao động
    e1 = []
    for data,actual in zip(X_test_manual, Y_test_manual):
        prediction = np.dot([data], beta)
        e1.append((actual-(prediction[0][0]))**2)
    RSS = np.sum(e1)
    RMSE = math.sqrt(RSS/n)
    return RMSE

# Test
print("Beta:", train_manual(X_train, Y_train))
print("RMSE:", test_manual(X_test, Y_test))
```
