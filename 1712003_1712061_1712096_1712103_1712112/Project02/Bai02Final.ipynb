```python
import numpy as np
import cv2 as cv
import csv
import math
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# đọc data từ tập train
def readData():
    X = []
    y = []
    with open('train.csv') as f:
        r = csv.reader(f)
        # Skip header
        next(r)
        # Đọc X, y
        for line in r:
            xline = []
            # Đọc từ 1 trở về sau
            for s in line[1:]:
                xline.append(float(s))
            X.append(xline)
            # Đọc y vào là cột
            y.append(float(line[0]))
    return (X,y)
# chuyển dòng thành ma trận 28x28
def convertmatrix(arr, i):
    a = arr[0][i]
    mat = []
    for x in range(0,len(a)):
        if (x % 28 == 0):
            mat.append([])
        mat[x//28].append(a[x])
    return mat
rData = readData()

# lọc nhiễu 
def gaussFilter(rData,i):
    data = convertmatrix(rData,i)
    alist = []
    ara = []
    X = []
    for i in data:
        alist = np.array(i)
        ara.append(alist)
    beta = np.array(ara)
    blur = cv.GaussianBlur(beta,(5,5),0.7)
    return blur

# tiền xử lí: xử lí ảnh và chuyển thành mảng dữ liệu X
def imageProcessing():
    X = []
    for i in range(27000):
        
        mat = gaussFilter(rData,i)
       
        xline = []
        for i in range(mat.shape[0]):
            for j in range(mat.shape[0]):
                xline.append(int(mat[i][j]))
        X.append(xline)
    return X
                             
X0 = imageProcessing()

# đọc nhãn
def readLabel():
    y = []
    with open('train.csv') as f:
        r = csv.reader(f)
        # Skip header
        next(r)
        # Đọc X, y
        for line in r:
            y.append(int(line[0]))
    return y
y0 = readLabel()


# đọc cả tập dữ liệu
X = np.array(X0[:])
y1 = np.array([y0[:]])
y = np.transpose(np.array(y0[:]))
print(len(y))
print(len(X))
```

```python
# xuất hình vị trí i (0<=i<=26999)
def showImage(i):
    data = convertmatrix(readData(),i)
    alist = []
    ara = []
    for i in data:
        alist = np.array(i)
        ara.append(alist)
    beta = np.array(ara)
    image = Image.fromarray(np.uint8(beta)).convert('RGB')
    image.save('myImage.png')
    image.show()
```

## Train và Test bằng Sklearn

```python
## Hàm train
import pandas as pd
from sklearn.linear_model import LogisticRegression

def train(train_x, train_y):
    ## Train cho mô hình hồi quy logistic đa thức
    clf = LogisticRegression(solver ='lbfgs',multi_class ='multinomial',max_iter = 20000)
    clf.fit(train_x,train_y)
    return clf
```

```python
## Chia tập dữ liệu (10% test)

#Có thể dùng đoạn code sau
##from sklearn.model_selection import train_test_split
##train_x, test_x, train_y, test_y = train_test_split(X,y,test_size = 0.1, random_state =0)
#
d1 = 2700
n = 27000-d1
train_x = np.array(X0[:n])
train_y = np.transpose(np.array(y0[:n]))
test_y = np.transpose(np.array(y0[n:]))
test_x = np.array(X0[n:])
## Gọi hàm train
model1 = train(train_x, train_y)
print("Done")
```

```python
## Hàm test kết quả
def test(test_x, test_y, clf):
    d = 0
    predicTest = clf.predict(test_x)
    for i in range(len(predicTest)):
        print("prediction: ",predicTest[i],"actual: ",test_y[i])
        if (predicTest[i] == test_y[i]):
            d+=1
    score = d/len(predicTest)
    return score
```

```python
##Dự đoán với 10% test
score = test(test_x, test_y, model1)
print("score: ",score)
model1.score(test_x,test_y)
```

```python
## 100% test_X train
full_train_X = X
full_train_y = y
model = train(full_train_X, full_train_y)
print("Done")
```

```python

print("score: ",test(test_x, test_y, model))
```
